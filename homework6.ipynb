```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#question 1.1
def sinusoid(x, A, w):
    return A*np.sin(w*x)
```

```python
#question 1.2
x_data = np.linspace(0,2*np.pi, 1000)
x_data
```

```python
#question 1.3
A_data = np.random.randint(low=-50, high=50, size=5)
w_data = np.random.randint(low=-2, high=2, size=5)
```

```python
#question 1.4
plt.figure()
for number, value in zip(A_data, w_data):
    y_data = sinusoid(x_data, number, value)
    plt.plot(x_data, y_data)
plt.title("5 Different Sine Functions")
plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.show()
```

```python
#question 2
firstlist = []
for number in np.random.randint(low=0, high=100, size=40):
    firstlist.append(number)
secondlist = []
for value in np.random.randint(low=0, high=100, size=40):
    secondlist.append(value)

x1_data = range(len(firstlist))

x2_data = range(len(secondlist))

plt.figure(figsize = (12,4))
plt.plot(x1_data, firstlist, color = 'orange', linewidth = 10)
plt.plot(x2_data, secondlist, color = 'red', linestyle = 'dashed')
plt.show()
```
